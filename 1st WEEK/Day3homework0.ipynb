FUNCTION/METHOD ALIASING

```python
def greet():
    print("Hello!")

# Create
say_hello = greet

# Call using
say_hello()  # Output: Hello!

#method

class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()

# Create an alias for the method
sum_func = calc.add

print(sum_func(3, 5))  # Output: 8
```
