##Filtering Data in a DataFrame

```python
#creating a distionary
data={
      "Name":["ram","shyam","hari"],
      "class":[7,8,9],
      "Roll":[1,2,3],
      "Address":["Pokhara","Butwal","Jhapa"]}
#turning distionary into a data frame
df=pd.DataFrame(data)
#Filtering
filter=df["Roll"]>1
print(df[filter])
filter=df["class"]>7
print(df[filter])
```

Reading a Processing a CSV fILE

```python
import io
csv_data="""
Name,Class,Roll,Address
ram,7,1,Pokhara
shyam,8,2,Butwal
hari,9,3,Jhapa     #Importing IO model for processing a CSv File
"""
df=pd.read_csv(io.StringIO(csv_data))   #using pandas & io attributes
print(df)
#df works data filter

#filtering the data
filter=df[df["Class"]>7]
filter=df[df["Roll"]>1]
print(filter)

#calculate the average grade
average=df["Class"].mean()
print(average)

```

Basic Plotting with Matplotlib

```python
from re import X
#import matplotlib as plt
import matplotlib.pyplot as plt

#defining daa for basic plotting
X=[1,2,3,4,5]
Y=[6,7,8,9,10]
plt.plot(X,Y,marker='o')  #mark the graph
plt.show()     #showing on graph
plt.xlabel("X-axis")   #define axis
plt.ylabel("Y-axis")
plt.show()

plt.grid(True)
plt.savefig("my_first_plot.png")
plt.show()






```

visualizing a Data Frame with Matplolib

```python
data={
      "Name":["ram","shyam","hari","sita"],
      "class":[7,8,9,10]
      }
df=pd.DataFrame(data)
plt.bar(df["Name"],df["class"])
plt.title("bar plot")
plt.xlabel("Name")
plt.ylabel("Class")

plt.show()
plt.savefig("my_second_plot.png")
```
