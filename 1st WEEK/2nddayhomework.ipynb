**Simple Interest Calculator**

```python
# principal amount
P = 1000

# rate of interest in percentage
R = 5  # Convert to decimal

# time in years
T = 2

# Use the formula: SI = P × R × T
simple_interest = (P * R * T)/100

# Print the result
print("Simple Interest is:", simple_interest)

```

**# BMI Calculator**


```python
#if
weight = 70       # in kg
height = 2        # in meters

# Calculate BMI
bmi= weight / (height ** 2)

# Print BMI
print(f"Your BMI is: {bmi:.2f}")

# Print weight category
if bmi < 15:
    print("You are underweight.")
elif bmi < 22:
    print("You are in the normal weight range.")
elif bmi < 28:
    print("You are overweight.")
else:
    print("You are obese.")
```

**Currency Converter**

```python
#Nepali to USD
 #Fixed exchange rate
exchange_rate = 132  # 1 USD = 132 NPR

# Amount in Nepali Rupees
NPR_amount = 5000

# Convert NPR to USD
usd_amount = NPR_amount / exchange_rate

# Display the result
print(f"{NPR_amount} NPR is equivalent to ${usd_amount:.2f} USD")
```

**SETS**

```python
# Add to set
fruits = {"apple","banana","mango","kiwi","grapes"}
fruits.add("watermelon")
print(fruits)
```

**Remove items on set**

```python
fruits = {"apple","banana","mango", "kiwi", "grapes"}
fruits.remove("banana")
print(fruits)
```

**Discard items on set**

```python
fruits = {"apple","banana","mango", "kiwi", "grapes"}
fruits.discard("papaya")
print(fruits)
```

**Clear set on set**

```python
fruits = {"apple","banana","mango", "kiwi", "grapes"}
fruits.clear()
print(fruits)
```

**Lenth on sets**

**TUPLE**

```python
# Define a list of student tuples
students =("kripa", 0, "K")
# Access an item by index
print("First student:",students[0])

# Length of the tuple
print("Total number of students:",len(students))

# Count how many times a value appears
print("Count of ('kripa', 0, 'K'):", students.count(("kripa")))

# Index of a specific student
print("Index of ('kripa', 1, 'K'):", students.index(("K")))

```
