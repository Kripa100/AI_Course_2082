**STRING MANUPULATION**

```python
#creating a variable named which  contain leading & trailing spaces
text="HELLO PYTHON"
print(text)
#using,lower () method /function to introduce lowecase
print(text.lower())
#using,strip ()
print(text.strip())

#Replace
print(text.replace("PYTHON","JAWA"))

#splitting & joinning
#method or function chaining
splitted=text.strip().split()
print(splitted)

 #using .join.() method to join the sepereted words
joined='-'.join(splitted)
print(joined)

#math
import math
squre_root=math.sqrt(25)
print("The squre root of 25 is",squre_root )

print(int(squre_root))

#flooring & Ceiling
floor=math.floor(4.8)
print(floor)

ceil=math.ceil(4.8)
print(floor)





```
